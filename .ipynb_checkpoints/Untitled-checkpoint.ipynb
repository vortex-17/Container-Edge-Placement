```python
#For graphs

import matplotlib.pyplot as plt
import numpy as np
from sklearn import preprocessing
import math
```

```python
total_nodes = np.asarray([30,70,90,150,200])
bestfit_nodes = np.asarray([27,63,88,140,184,333,542,607,1013,2026,2535])
custom_nodes = np.asarray([27,62,87,138,182])

# normalised_bestfit = preprocessing.normalize(bestfit_nodes, axis = 0)

norm1 = total_nodes / np.linalg.norm(total_nodes)
norm2 = (bestfit_nodes / np.linalg.norm(bestfit_nodes))
norm3 = (custom_nodes / np.linalg.norm(custom_nodes)) 
print(norm2)

plt.plot(norm1, norm2, color='green', linestyle='dashed', linewidth = 1,
         marker='o', markerfacecolor='blue', markersize=1)
plt.plot(norm1, norm3)
plt.show
```

```python
x = np.arange(5)
plt.bar(x, norm2, color = "black")
plt.bar(x + 0.1 , norm3, color = "red")
plt.xticks(x, ['Team A', 'Team B', 'Team C', 'Team D', 'Team E'])
plt.show()
```
